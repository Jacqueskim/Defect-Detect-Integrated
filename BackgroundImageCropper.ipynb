```python
directoryPath = "/content/shareddrive/MyDrive/CE_Capstone_QA_2022-2023/DataCreation"
readPath = "uncutVALIDATION"
writePath = "cutVALIDATION"

os.makedirs(os.path.join(directoryPath,writePath), exist_ok=True)
startDim = (300,224)
endDim = (224,168)
```

```python
def loadAndCropImages(path):
  os.chdir(path)
  images = []
  for image_file in glob.glob("*.png") + glob.glob("*.jpg"):
    # Read the image using OpenCV
    img = cv2.imread(image_file)
   
    img = img[0:endDim[0], startDim[1]-endDim[1]:startDim[1]]
    #img = cv2.resize(img)
    images.append(img)
   # cv2_imshow(img)
    #print(img.shape)
  return images
```

```python
images = loadAndCropImages(os.path.join(directoryPath,readPath))


```

```python
for i in range(10):
  cv2_imshow(images[i])
  print(images[i].shape)
```

```python
# Loop over the image list
for i, image in enumerate(images):
    # Read the image using OpenCV
    

    # Create the output filename
    filename = f'image{i}.jpg'

    # Create the output file path
    output_path = os.path.join(directoryPath, writePath)
    output_path = os.path.join(output_path,filename)

    # Write the image to the output folder
    cv2.imwrite(output_path, image)

    # Print a message to indicate the progress
    print(f"Image {filename} saved to {output_path}")
```
